```python
from pyspark.sql import SparkSession
import getpass
username = getpass.getuser()
spark = SparkSession. \
builder. \
config('spark.ui.port', '0'). \
config("spark.sql.warehouse.dir", f"/user/{username}/warehouse"). \
config('spark.shuffle.useOldFetchProtocol', 'true'). \
config('spark.app.name', 'itv012041_Lending_Club'). \
enableHiveSupport(). \
master('yarn'). \
getOrCreate()
```

```python
raw_df = spark.read \
.format("csv") \
.option("inferSchema", "true") \
.option("header", "true") \
.load("/public/trendytech/datasets/accepted_2007_to_2018Q4.csv")
```

```python
raw_df.createOrReplaceTempView("lending_club_data")
```

```python
spark.sql("select * from lending_club_data")
```

```python
from pyspark.sql.functions import sha2, concat_ws
```

```python
new_df = raw_df. \
withColumn("member_id_hashed",sha2(concat_ws("+",*["emp_title","emp_length","home_ownership","annual_inc","zip_code","addr_state","grade","sub_grade","verification_status"]),256))
```

```python
new_df
```

```python
new_df.createOrReplaceTempView("lending_club_data_v2")
```

```python
spark.sql("select count(*) from lending_club_data_v2")
```

```python
spark.sql("select count(distinct(member_id_hashed)) as distinct_member_id from lending_club_data_v2")
```

```python
spark.sql("""select member_id_hashed, count(*) as total_count
from lending_club_data_v2
group by 1
having total_count > 1
order by 2 desc""")
```

```python
spark.sql("select * from lending_club_data_v2 where member_id_hashed like 'e3b0c44298fc1c149%'")
```

```python
spark.sql("""
select member_id_hashed as member_id,emp_title,emp_length,home_ownership,annual_inc,addr_state,zip_code,
'USA' as country,grade,sub_grade,verification_status,tot_hi_cred_lim,application_type,
annual_inc_joint,verification_status_joint
from lending_club_data_v2
""").repartition(1).write \
.option("header", True) \
.format("csv") \
.mode("overwrite") \
.option("path", "/user/itv012041/lendingclubproject/raw/customers_data_csv") \
.save()
```
